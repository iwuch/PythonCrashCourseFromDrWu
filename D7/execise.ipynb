```python
import pandas as pd
#通过read_csv函数来读取to_read.csv文件中的数据并且默认以DataFrame的格式存儲
to_read = pd.read_csv("to_read.csv")
#应用value_counts函数按照book_id的出现次数进行计数，并且使用sort_values对齐进行降序
to_read_value_count = to_read['book_id'].value_counts().sort_values(ascending=False)
#通过切片索引选择排序之后的前50个本书，生成一个Series数据结构
hottest_top_50_books = to_read_value_count[:50]
#提取数据的index和values，来创建新的DataFrame数据
hottest_50_books_id_col = hottest_top_50_books.index
hottest_50_books_counts = hottest_top_50_books.values
to_read_counts_df = pd.DataFrame({
        'book_id':hottest_50_books_id_col,
        'counts':hottest_50_books_counts
    })
#读取books.csv，提取其中的'book_id','goodreads_book_id','title'三列与DataFrame进行合并
books = pd.read_csv('books.csv',encoding="ISO-8859-1")
books_id_and_title = books[['book_id','goodreads_book_id','title']]
hottest_books_list = pd.merge(to_read_counts_df,books_id_and_title,on ='book_id',how='left')
print(hottest_books_list)

#编写一个函数，在data数据表(DataFrame格式)中找到goodreads_book_id这本书对应第N热门的那个标签

N=0
goodreads_book_id=18143977
book_tags = pd.read_csv("book_tags.csv",index_col="_goodreads_book_id_")
tags = pd.read_csv("tags.csv")

def find_Nth_tag_of_the_book(goodreads_book_id,N,book_tags):
    data = book_tags[book_tags.index==goodreads_book_id].sort_values('count',ascending=False)
    tag_id = data.iloc[N].tag_id
    tag_name = tags[tags.tag_id==tag_id].tag_name.values[0]
    return tag_name
#使用for循环遍历10个标签和50本书的，每个标签作为一列插入到hottest_books_list的最后
for N in range(10):
    column_name = "Tag"+str(N+1)
    tag_list = []
    for goodreads_book_id in hottest_books_list.goodreads_book_id:
        tag_list.append(find_Nth_tag_of_the_book(goodreads_book_id,N,book_tags))
    new_Series = pd.Series(tag_list)
    hottest_books_list[column_name]=new_Series
print(hottest_books_list)

hottest_books_list.to_csv("hottest_books_list.csv",index=False,encoding="utf8")
```
